```
import pandas as pd

df = pd.read_csv('https://query.data.world/s/44dbtk644cvxiciot5fiuzserogpkv')

df.head()
```

```
years = dict()

for i in range(len(df['SongID'])):
  year = df['WeekID'][i]
  year = year[len(year)-4:]

  songID = df['SongID'][i]
  songName = df['Song'][i]
  artist = df['Performer'][i]
  if year not in years:
    years[year] = dict()
    years[year][songID] = [songName, artist, None]
  elif songID not in years[year]:
    years[year][songID] = [songName, artist, None]
```

```
import collections

years = collections.OrderedDict(sorted(years.items()))
years.keys()
```

```
from google.colab import files

lyrics_df = pd.DataFrame(columns=['Year', 'SongID', 'Lyrics'])

def make_dataset():
  global lyrics_df
  # count = 0
  for y in range(2000, 2019):
    year = str(y) 
    lyrics = dict()
    for songID in years[year]:
      # count+=1
      # if count>20:
      #   break
      if songID not in lyrics:
        try:
          lyrics[songID] = get_lyrics(songID)
        except:
          pass
        try:
          row = pd.DataFrame({'Year':[year], 'SongID':[songID], 'Lyrics':[lyrics[songID]]})
          lyrics_df = pd.concat([lyrics_df, row])
        except:
          pass
      
         

make_dataset()

lyrics_df.to_csv('lyrics_df.csv')
files.download('lyrics_df.csv')
```
