```
from google.colab import files

lyrics = dict()
lyrics_df = pd.DataFrame(columns=['Artist', 'Song', 'SongID', 'Lyrics', 'Label'])

def make_dataset():
  global lyrics_df
  for i in range(len(df['Index'])):
    songID = df['Title'][i] + df['Artist'][i]
    if songID not in lyrics:
      try:
        lyrics[songID] = get_lyrics(songID)
      except:
        pass
      try:
        row = pd.DataFrame({'Artist':df['Artist'][i], 'Song': df['Title'][i], 'SongID':[songID], 'Lyrics':[lyrics[songID]], 'Label': df['Mood'][i]})
        lyrics_df = pd.concat([lyrics_df, row])
      except:
        pass

make_dataset()

lyrics_df.to_csv('lyrics_df.csv')
files.download('lyrics_df.csv')
```
