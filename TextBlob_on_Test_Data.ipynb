```
from textblob import TextBlob

def sentiment_scores(lyrics):
  score = 0
  sentences = lyrics.split('\n')
  for sentence in sentences: 
    blob = TextBlob(sentence) 
    for text in blob.sentences: 
      score += text.sentiment.polarity
    
  return score
    
```

```
from google.colab import files 
  
uploaded = files.upload()
```

```
import pandas as pd 
import io 
  
df = pd.read_csv(io.BytesIO(uploaded['lyrics_df.csv'])) 
df = df.drop(columns=['Unnamed: 0'])
```

```
df.head(10)
```

```
pred_df = pd.DataFrame(columns=['Artist', 'Song', 'Label', 'Predicted Label'])

def predict_labels():
  global pred_df
  for i in range(len(df['Lyrics'])):
    if type(df['Lyrics'][i]) != str:
      continue
    score = sentiment_scores(df['Lyrics'][i])
    pred_label = 'neg' if score<=0 else 'pos'
    row = pd.DataFrame({'Artist': df['Artist'][i], 'Song': df['Song'][i], 'Label': df['Label'][i], 'Predicted Label': [pred_label]})
    pred_df = pd.concat([pred_df, row])
    print(len(df['Lyrics'])-i, "left")
```

```
predict_labels()
```
